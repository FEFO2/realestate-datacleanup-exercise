```python
# (1) created a function that shows all columns and the amount of NAs that they have. 
def printNAs(datfr):
    datfr = datfr.isna()
    cols = []
    trues = []
    falses = []
    for x in range(0,len(datfr.axes[1])):
        cols.append(list(datfr.columns)[x])
        trues.append(datfr.iloc[:,x].sum())
        falses.append(len(datfr.iloc[:,x])-datfr.iloc[:,x].sum())
    data = {
    "cols": cols,
    "trues": trues,
    "falses": falses
    }
    result = pd.DataFrame(data)
    print(result)
printNAs(new_ds)

# after seeing the results, I decided to drop column 35 and 36 (only NAs)

print("new_ds.shape = {}".format(new_ds.shape))
clean_ds = new_ds.drop(['zipCode','customZone'], axis=1)
print("clean_ds.shape = {}".format(clean_ds.shape))


```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# FUNCTION TO NORMALIZE
def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)
    for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return norm_arr

#NORMALIZATION OF PRICES PER POPULATION
nor_price_f18 = normalize(nds_madrid.loc[nds_madrid['level5']=='Fuenlabrada']['price'].tolist(),0,1)
nor_price_l18 = normalize(nds_madrid.loc[nds_madrid['level5']=='Leganés']['price'].tolist(),0,1)
nor_price_g18 = normalize(nds_madrid.loc[nds_madrid['level5']=='Getafe']['price'].tolist(),0,1)
nor_price_a18 = normalize(nds_madrid.loc[nds_madrid['level5']=='Alcorcón']['price'].tolist(),0,1)

#MULTIHIST PLOT
colo = ['red', 'blue', 'lime','orange']
names = ['Fuenlabrada','Leganés','Getafe','Alcorcón']
fig, hisplot = plt.subplots(2,2)
fig.suptitle('Comparison of normalized prices of houses , South belt Madrid')
hisplot[0,0].hist(nor_price_f18, bins = 30, density=True, histtype='bar')
hisplot[0,0].set_title('Fuenlabrada')
hisplot[0,1].hist(nor_price_l18, bins = 30, density=True, histtype='bar',color='green')
hisplot[0,1].set_title('Leganés')
hisplot[1,0].hist(nor_price_g18, bins = 30, density=True, histtype='bar',color='red')
hisplot[1,0].set_title('Getafe')
hisplot[1,1].hist(nor_price_a18, bins = 30, density=True, histtype='bar',color='orange')
hisplot[1,1].set_title('Alcorcón')

for ax in hisplot.flat:
    ax.set(xlabel='Surface (mts2)', ylabel='Price (EUR)')
for ax in hisplot.flat:
    ax.label_outer()
```

#This multihist shows the media of the prices of the four populations. Using normalized prices allow the graph to distribute the results over a wider range than the original prices. Also, it allows us to compare the prices of the houses in the same range of a scale.

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in the question 14) (★★★★)

Execute the following cell and then start coding in the next one. You must implement a simple code that transform the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
# (1) import the library and create the map
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon

map = Map(center=(60, -2.2), zoom=2, min_zoom=1, max_zoom=20,
           basemap=basemaps.Stamen.Terrain)

# (2) Define a dictionary to map values to marker colors
color_dict = {"Fuenlabrada": "blue", "Leganés": "green", "Getafe": "red", "Alcorcón": "purple"}

# (3) filter out the 0 lat/long from the dataframe 
nds_madrid2 = nds_madrid[(nds_madrid['latitude'] != '0') &
                         (nds_madrid['longitude'] != '0')]

# Use list comprehension to create tuples from latitude and longitude columns
latitude= [float(lat.replace(',', '.')) for lat in nds_madrid2['latitude']]
longitude= [float(long.replace(',', '.')) for long in nds_madrid2['longitude']]

# Loop through the latitude and longitude lists and add markers to the map
for i, (lat, long) in enumerate(zip(latitude, longitude)):
    population = nds_madrid2.iloc[i]['level5']
    color = color_dict.get(population, "black")
    marker = Marker(location=(lat, long), icon=AwesomeIcon(name='house', marker_color=color), draggable=False)
    map.add_layer(marker)

map

```
